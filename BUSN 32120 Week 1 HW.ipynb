```python
def fizzbuzz(n) :
    n = int(n)
    for n in range (1, n+1)
        if n % 3 == 0 and n % 5 == 0: 
            print ("fizzbuzz")
        elif n % 3 == 0:
            print ("fizz")
        elif n % 5 == 0:
            print ("buzz")
        else:
            print(n)

# Identification
# Variable: n - names that hold data in a computer program
# Loop: for n in range - repeats a piece of code a set number of times
# Conditional: if n % 5 - executes different actions based on whether a condition is true or false
# Return: print(n) - ends execution and returns control back to the code that called the function
```
