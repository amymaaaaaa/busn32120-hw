## 5. TB Dataset

### Load TB Dataset

Load the dataset and examine its structure to understand how the variables are stored.

```python
tb = pd.read_csv("TB_notifications_LK_edit.csv")
tb.head()
```

```python
tb.info()
```

```python
tb.shape
```

```python
tb.columns
```

### Convert Data using melt

The dataset stores case counts in many columns where gender and age group are embedded in the column names. The dataset was reshaped into long format so that each row represents a single observation.

```python
tb_tidy = tb.melt(
    id_vars=["country", "iso3", "iso_numeric", "whoregion and year"],
    var_name="case_type",
    value_name="cases"
)
```

```python
tb_tidy = tb_tidy.dropna(subset=["cases"])
```

### Extract Gender and Age Group

The column names contain multiple variables (case type, gender, and age group). Extract these components into separate columns.

```python
tb_tidy["gender"] = tb_tidy["case_type"].str.extract(r'_(m|f)')
```

```python
tb_tidy["age_group"] = tb_tidy["case_type"].str.extract(r'(\d+)')
```

```python
tb_tidy["case_category"] = tb_tidy["case_type"].str.extract(r'(new_[a-z]+)')
```

```python
tb_tidy = tb_tidy.drop(columns=["case_type"])
```

```python
tb_tidy = tb_tidy[
    ["country", "iso3", "iso_numeric", "whoregion and year",
     "case_category", "gender", "age_group", "cases"]
]
```

```python
tb_tidy = tb_tidy.sort_values(
    ["country", "whoregion and year", "gender", "age_group"]
)
```

```python
tb_tidy = tb_tidy.dropna(subset=["gender", "age_group"])
```
